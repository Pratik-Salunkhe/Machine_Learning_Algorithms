# CROSS-TABULATION ON `h1n1_vaccine_prediction`

###### - Importing Pandas Library

```python
import pandas as pd
```

###### - Reading csv File

```python
v = pd.read_csv("h1n1_vaccine_prediction.csv")
v
```

```python
v.shape
```

### CROSS TABULATION

```python
pd.crosstab(v.income_level,v.h1n1_vaccine)
```

```python
print(10178/(10178+2599)*100,"% of people has not taken a vaccine whose income level is Above Poverty")
print(2599/(10178+2599)*100, "% of people has taken a vaccine whose income level is Above Poverty","\n"," ")

print(5087/(5087+1723)*100,"% of people has not taken a vaccine whose income level is greater than 75,000 $")
print(1723/(1723+5087)*100, "% of people has taken a vaccine whose income level is greater than 75,000 $","\n"," ")

print(5087/(5087+1723)*100,"% of people has not taken a vaccine whose income level is Below Poverty")
print(1723/(1723+5087)*100, "% of people has taken a vaccine whose income level is Below Poverty","\n"," ")
```

```python
pd.crosstab(v.marital_status,v.h1n1_vaccine)
```

```python
print(10387/(10387+3168)*100,"% of peoples are married who has not taken a vaccine")
print(3168/(10387+3168)*100, "% of peoples are married who has taken a vaccine","\n"," ")

print(9494/(9494+2250)*100,"% of peoples are not married who has not taken a vaccine")
print(2250/(9494+2250)*100, "% of peoples are not married who has taken a vaccine","\n"," ")
```

```python
pd.crosstab(v.sex,v.h1n1_vaccine)
```

###### - Cross-Tabulation on all columns

```python
x = v.iloc[:,1:-1]
y = v.iloc[:,-1]
col = x.select_dtypes(include = 'object').columns

for i in col:
    print("Cross Tabulation on ",i,"& h1n1_vaccine")
    print("")
    print(pd.crosstab(x[i],y))
    print("")
```
