# DATA CLEANING ON `Patient_with_abnormal_bloodpressure.csv` DATASET

###### - Importing pandas Library

```python
import pandas as pd
```

###### - Reading a csv File

```python
bcc = pd.read_csv("Patient_with_abnormal_bloodpressure.csv")
bcc
```

### Handeling Of Nulls

###### - Checking for the columns contaning Null values

```python
bcc.isnull().sum()
```

###### - Filling the values in place of Nulls

```python
bcc.alcohol_consumption_per_day.value_counts()
```

```python
# bcc.Credit_Product.fillna("No", inplace = True)

bcc.Genetic_Pedigree_Coefficient.fillna(bcc.Genetic_Pedigree_Coefficient.mean() , inplace = True)
bcc.Pregnancy.fillna(0 , inplace = True)
bcc.alcohol_consumption_per_day.fillna(0 , inplace = True)
```

###### - Checking for columns containing Null values after filling of values in place of Null

```python
bcc.isnull().sum()[bcc.isnull().sum() > 0]
```

### Droping Of Meaningless / Unwanted Columns

###### - Droping the "Patient_Number" column from dataset

```python
bcc.drop(columns = ["Patient_Number"], axis = 1, inplace = True)
bcc.head()
```
