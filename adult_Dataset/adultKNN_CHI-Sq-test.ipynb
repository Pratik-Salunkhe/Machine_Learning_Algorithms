### CHI-Square Test

```python
def chi_sq_test (input_tabel):
    from scipy.stats import chi2_contingency as chi2_c
    pvalue = chi2_c(input_tabel)[1]
    if pvalue < 0.05:
        print("Pvalue = ", pvalue)
        print("Reject the null as Pvalue is less than 0.05")
        print("Conclusion :  There is a relationship between 2 variables")
    else:
         print("Pvalue = ", pvalue)
         print("Fail to reject the null as Pvalue is greater than 0.05")
         print("Conclusion :  There is No relationship between 2 variables")
```

```python
chi_sq_test(pd.crosstab(ak.workclass,ak.income))
```
